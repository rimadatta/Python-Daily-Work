##### Plotting an array as image

```python
import numpy as np
```

```python
img_new=np.zeros([100,100])
```

```python
plt.imshow(img_new, cmap='gray') #colormap=gray, all zeros means all black pixels
```

```python
img_new[34,57]=0.5 #individual pixel value changed
```

```python
plt.imshow(img_new, cmap='gray')
```

```python
img_new[34]=1 # whole row is 1
plt.imshow(img_new, cmap='gray')
```

```python
np.random.randint(12,78)
```

```python
img_new=np.zeros([100,100],dtype=np.uint8)
```

```python
for i in range(100):
    for j in range(100):
        img_new[i,j]=np.random.randint(0,255)
```

```python
plt.imshow(img_new,cmap='grey')
```

```python
#using float
import random
img_new=np.zeros([100,100],dtype=np.float16)
for i in range(100):
    for j in range(100):
        img_new[i,j]=random.random()
plt.imshow(img_new,cmap='grey')
```

##### Greyscale image

```python
img_grey=cv2.imread("images/lena.png",0) #the second parameter is used to mention that we are using 2 channels of greyscale
```

```python
img_grey.shape
```

```python
plt.imshow(img_grey, cmap='grey')
```

```python
img_plt.shape
```

```python
#Slicing img_plt
r=img_plt[:,:,0]
g=img_plt[:,:,1]
b=img_plt[:,:,2]
```

```python
r.shape, g.shape, b.shape
```

```python
plt.subplot(1,3,1) # function to plot 3 simultaneous plots
plt.imshow(r,cmap='grey')
plt.subplot(1,3,2)
plt.imshow(g,cmap='grey')
plt.subplot(1,3,3)
plt.imshow(b,cmap='grey')
```

```python
plt.subplot(1,3,1) # function to plot 3 simultaneous plots
plt.imshow(r,cmap='Reds')
plt.subplot(1,3,2)
plt.imshow(g,cmap='Greens')
plt.subplot(1,3,3)
plt.imshow(b,cmap='Blues')
```

```python
plt.imshow(img_plt) #after combining all RGB, we get original image
```

```python
#Grey image is the average of RGB
new_img=(r+g+b)/3
plt.imshow(new_img, cmap='gray')
```

```python
plt.subplot(1,2,1)
plt.title('Image read using CV2')
plt.imshow(img1) #Internally CV2 library uses BGR channel
plt.subplot(1,2,2)
plt.title('Image read using matplotlib') #Internally Matplotlib library uses RGB channel
plt.imshow(img_plt)
```

```python
img2=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB) #CONVERTING BGR TO RGB PATTERN EXPLICITLY
```

```python
plt.subplot(1,2,1)
plt.title('Image read using CV2')
plt.imshow(img2) #Internally CV2 library uses BGR channel
plt.subplot(1,2,2)
plt.title('Image read using matplotlib') #Internally Matplotlib library uses RGB channel
plt.imshow(img_plt)
```
