```python
import cv2
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('images/dog.jpg')
image =cv2.cvtColor(image,cv2.COLOR_BGR2RGB)

#Flip the image horizontally
flipped_horizontal = cv2.flip(image,0)
#flipped_horizontal=cv2.cvtColor(flipped_horizontal,cv2.COLOR_BGR2RGB)

#Flip the image vertically
flipped_vertical = cv2.flip(image,1)
#flipped_vertical=cv2.cvtColor(flipped_vertical,cv2.COLOR_BGR2RGB)

# Flip the image diagonally
flipped_diagonal = cv2.flip(image,-1)
#flipped_diagonal=cv2.cvtColor(flipped_diagonal,cv2.COLOR_BGR2RGB)

#Show the flipped images
plt.figure(figsize=(16,9))
plt.subplot(1,4,1)
plt.title('Original Image')
plt.imshow(image)
plt.subplot(1,4,2)
plt.title('Horizontally Flipped Image')
plt.imshow(flipped_horizontal)
plt.subplot(1,4,3)
plt.title('Vertically Flipped Image')
plt.imshow(flipped_vertical)
plt.subplot(1,4,4)
plt.title('Diagonally Flipped Image')
plt.imshow(flipped_diagonal)
```
