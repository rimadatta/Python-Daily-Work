##### Bitwise AND

```python
import cv2
import matplotlib.pyplot as plt
image1 = cv2.imread("images/input1.png")
image2 = cv2.imread("images/input2.png")
```

##### Bitwise AND

```python
dest_and=cv2.bitwise_and(image1, image2, mask = None)
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,3,1)
plt.imshow(image1)
plt.subplot(1,3,2)
plt.imshow(image2)
plt.subplot(1,3,3)
plt.imshow(dest_and)
```

##### Bitwise OR

```python
dest_or=cv2.bitwise_or(image1, image2, mask = None)
plt.figure(figsize=(16,9))
plt.subplot(1,3,1)
plt.imshow(image1)
plt.subplot(1,3,2)
plt.imshow(image2)
plt.subplot(1,3,3)
plt.imshow(dest_or)
```

##### Bitwise NOT

```python
dest_not = cv2.bitwise_not(dest_or, mask = None)
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.imshow(dest_or)
plt.subplot(1,2,2)
plt.imshow(dest_not)
```

##### Bitwise XOR

```python
dest_xor = cv2.bitwise_xor(image1,image2, mask = None)
plt.figure(figsize=(16,9))
plt.subplot(1,3,1)
plt.imshow(image1)
plt.subplot(1,3,2)
plt.imshow(image2)
plt.subplot(1,3,3)
plt.imshow(dest_xor)
```
