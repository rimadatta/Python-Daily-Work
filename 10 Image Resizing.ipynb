```python
import cv2

#Load the image
image=cv2.imread("images/lena.png")

# Resize the image to a fixed width and height
resized_image=cv2.resize(image,(100,200))
original_image=cv2.resize(image,(500,500))

#show the resized image
cv2.imshow("Resized Image",resized_image)
cv2.imshow("Original Image",original_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

##### Rescaling Images

```python
import cv2

#Load the image
image=cv2.imread("images/lena.png")

# Rescale the image to a fixed width and height
resized_image1=cv2.resize(image,None, fx=0.5, fy=0.75) # Aspect ratio not preserved
resized_image2=cv2.resize(image,None, fx=1.5, fy=1.5)  # Aspect ratio is preserved
original_image=cv2.resize(image,(500,500))

#show the resized image
cv2.imshow("Resized Image1",resized_image1)
cv2.imshow("Resized Image2",resized_image2)
cv2.imshow("Original Image",original_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

##### Resize using Skimage

```python
from skimage.transform import resize
import matplotlib.pyplot as plt

#Load the image
image = plt.imread('images/dog.jpg')

#Resize with bilinear interpolation
resized_image = resize(image, (1300,1300), anti_aliasing=True)

#Resize with bicubic interpolation
#resized_image = resize(image, (1300,1300), anti_aliasing = True, order =3)

#Show the resized image

plt.subplot(1,2,1)
plt.imshow(image)
plt.subplot(1,2,2)
plt.imshow(resized_image)
```
