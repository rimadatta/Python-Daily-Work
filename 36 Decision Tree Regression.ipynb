```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pdf = pd.read_csv('Dataset/petrol_consumption.csv')
pdf
```

```python
pdf.shape
```

```python
# Output variable is petrol consumption. It is not categorical. Hence we use Decision Tree Regression
```

##### Separate input and output data

```python
x = pdf.drop('Petrol_Consumption', axis =1)
y= pdf['Petrol_Consumption']
```

```python
x.corrwith(y) # correlation between input and output is very less on this data, that's why we don't use linear regression
```

```python
sns.kdeplot(y)
```

##### Cross validation

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state =0)
```

```python
x_train.shape, x_test.shape
```

##### Build the model

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
regressor = DecisionTreeRegressor(random_state = 0)
```

```python
regressor.fit(x_train, y_train)
```

##### Evaluate the performance

```python
y_pred = regressor.predict(x_test)
```

```python
result = pd.DataFrame({'Actual': y_test, 'Predicted':y_pred})
result
```

```python
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, mean_squared_error
```

```python
mean_absolute_error(y_test, y_pred) # R2 Error is not used in regression. It is for classification only
```

```python
mean_absolute_percentage_error(y_test, y_pred) # almost 13% error
```

```python
np.sqrt(mean_squared_error(y_test, y_pred))
```

##### Visualize

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(16,9))
plot_tree(regressor, fontsize = 6, feature_names=x.columns, filled = True, precision = 2, rounded = True);
```

```python
# To print decision tree gives importance to which importance
plt.title('Feature Importances')
plt.xlabel('Features')
plt.ylabel('Importances')
plt.bar(x.columns, regressor.feature_importances_, color = 'purple')
```

```python
new = pd.DataFrame(data=[[7.3, 4449, 4639, 0.626],
                         [6.4, 5215, 2302, 0.672]], columns = x.columns)
```
