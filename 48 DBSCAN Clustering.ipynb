```python
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
```

##### Create the dataset

```python
# Step 1:Create a sample dataset
x, y = make_blobs(n_samples = 300, centers = 4, cluster_std = 0.60, random_state = 0) # We can pass centers,standard deviation,
```

```python
y # Output of data
```

```python
x # data
```

```python
x.shape
```

```python
plt.xlabel('Feature-1')
plt.ylabel('Feature-2')
plt.scatter(x[:,0],x[:,1])
# see 4 clusters created
```

 ##### Create the object

```python
# Step 2: Apply DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples =5) # eps is Epsilon value
clusters = dbscan.fit_predict(x)
```

```python
clusters # -1 are noise which are not within the scope of any eps value
```

```python
plt.xlabel('Feature-1')
plt.ylabel('Feature-2')
plt.scatter(x[:,0],x[:,1], c=clusters) #The points in clusters will change on changing epsilon value. The purple points are currently noise with eps=0.5
```
