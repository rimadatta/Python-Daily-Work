```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
num = [np.random.randint(0,100) for x in range (1000)] # generates random no between 1-100 and save it in an array of size 1000
```

```python
num
```

```python
plt.title('Histograms')
plt.xlabel('Values')
plt.ylabel('Count')
plt.xticks(range(10,101,10))
plt.grid()
plt.hist(num, bins=range(10,101,10))
```
