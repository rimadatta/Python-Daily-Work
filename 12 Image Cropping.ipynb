```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
img=plt.imread('images/lena.png')
```

```python
image=img[150:300,200:350]
```

```python
plt.subplot(1,2,1)
plt.imshow(img)
plt.subplot(1,2,2)
plt.imshow(image)
```

```python
height, width, channels = img.shape
```

```python
# Delete 15% of image from all the sides of the image
startrow = int(height*0.15)
startcol = int(width*0.15)
endrow = int(height*0.85)
endcol = int(width*0.85)
```

```python
croppedimage = img[startrow:endrow, startcol:endcol]
```

```python
plt.subplot(1,2,1)
plt.imshow(img)
plt.subplot(1,2,2)
plt.imshow(croppedimage)
```
