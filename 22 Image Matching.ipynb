```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
image1 = cv2.imread('images/eiphel1.png')
image2 = cv2.imread('images/eiphel2.png')
```

```python
sift = cv2.SIFT_create()
```

```python
keypoints_1, descriptors_1 = sift.detectAndCompute(image1, None) # main pixels in an image which the program keeps note of
keypoints_2, descriptors_2 = sift.detectAndCompute(image2, None)

len(keypoints_1), len(keypoints_2) # will print number of key pixels in image 1 and image 2
```

```python
# feature Matching
bf = cv2.BFMatcher(cv2.NORM_L1)

matches = bf.match(descriptors_1, descriptors_2)
matches = sorted(matches, key = lambda x:x.distance)   # lambda is an anonymous function in core python where distance between 2 points is calculated and sorted

img3 = cv2.drawMatches(image1, keypoints_1, image2, keypoints_2, matches[:40], image2, flags=2) #will show 1st 40 matches using keypoints in image 1 and image and draw lines. The way of representation is way no 2 in flags
plt.imshow(img3)
```
