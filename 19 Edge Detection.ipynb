```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread('images/lena.png',0)
# For edge detection, data is cumpulsarily required in greyscale
```

##### Sobel Edge Detection

```python
# Sobel Parameters : ksize, dx, dy, ddepth
sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3) # Sobel for horizontal edges, cv2.CV_64F is a constant for depth
sobely = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3) # Sobel for vertical edges

# Combine the Sobel Images (optional)
combined_sobel = cv2.add(sobelx, sobely)
```

```python
final_img=np.hstack([sobelx, sobely, combined_sobel])
cv2.imshow('All Images', final_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.figure(figsize=(16,9))
plt.subplot(141)
plt.title('Original Image')
plt.imshow(img, cmap='gray')
plt.subplot(142)
plt.title('Sobel X')
plt.imshow(abs(sobelx), cmap='gray') #abs function is used to convert negative values in kernel to positive values
plt.subplot(143)
plt.title('Sobel Y')
plt.imshow(abs(sobely), cmap='gray')
plt.subplot(144)
plt.title('Combined')
plt.imshow(abs(combined_sobel), cmap='gray')
```

```python
# Load the image
img = cv2.imread('images/circle.png',0)

# Apply Canny Edge detection
edges = cv2.Canny(img, 50, 150) #adjust thresholds if needed

# Display results

cv2.imshow('Original Image', img)
cv2.imshow('Canny Edge Detection', edges)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
