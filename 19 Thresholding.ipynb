```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

##### performing manual threshold with value 127

```python
img=cv2.imread('images/lenna.png',0)
new_img = img>127 # keeps all the pixels greater than 127 as white, rest all pixels are converted to black(false). Genrally background pixels has higher intensity so they are converted to white
new_img
```

```python
inv_img = img<127
inv_img
```

```python
plt.figure(figsize=(16,9))
plt.subplot(131)
plt.title('Original Image')
plt.imshow(img, cmap='gray')
plt.subplot(132)
plt.title('Binary Image')
plt.imshow(new_img, cmap='gray')
plt.subplot(133)
plt.title('Inverted Binary Image')
plt.imshow(inv_img, cmap='gray')
```

##### Otsu Threshold

```python
# Apply Otsu Threshold
thresh_val, thresh_otsu = cv2.threshold(img, 0, 255, cv2.THRESH_OTSU) # Return value that has highest between class variance calculated internally
```

```python
thresh_val
```

```python
plt.figure(figsize=(16,9))
plt.subplot(141)
plt.title('Original Image')
plt.imshow(img, cmap='gray')
plt.subplot(142)
plt.title('Binary Image')
plt.imshow(new_img, cmap='gray')
plt.subplot(143)
plt.title('Inverted Binary Image')
plt.imshow(inv_img, cmap='gray')
plt.subplot(144)
plt.title('Otsu Binary Image')
plt.imshow(thresh_otsu, cmap='gray')
```
