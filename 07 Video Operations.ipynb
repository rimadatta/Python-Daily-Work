```python
#Metadata
width=int(vid.get(cv2.CAP_PROP_FRAME_WIDTH))
height=int(vid.get(cv2.CAP_PROP_FRAME_HEIGHT))
frames=int(vid.get(cv2.CAP_PROP_FRAME_COUNT))
fps=int(vid.get(cv2.CAP_PROP_FPS))
```

```python
print("Video Resolution: ",height,width)
print("Total frames: ",frames)
print("FPS:",fps)
```

```python
#play video by extracting frames from video robots
vid=cv2.VideoCapture("images/robots.mp4")
while True:
    ret, frame=vid.read()
    if not ret:
        break
    #Process the frame (e.g., apply filters, detect objects)
    cv2.imshow("Video",frame)
    if cv2.waitKey(15)==ord('q'):
        break
cv2.destroyAllWindows()
```

```python
#save all frame in video as pictures
vid=cv2.VideoCapture("images/robots.mp4")
#get the first frame and see if successful
success, frame=vid.read()
#initiate count
count=1
#keep extracting frames as long as success is True
while success:
    cv2.imwrite(f'Video_Frames/{count}.jpg',frame)
    success, frame=vid.read()
    count+=1
```

##### Extracting particular frame from video

```python
timestamp='00:00:43'
timestamp.split(':')
```

```python
# get frame at a speific timestamp by separating hours, minutes and seconds
timestamp='00:00:43.00'
hours,minutes,seconds=[float(i) for i in timestamp.split(':')]
print(hours,minutes,seconds)
```

```python
fps
```

```python
#get number of frames upto timestamp
trigger_frame = hours*3600*fps + minutes*60*fps + seconds*fps
print(frames, trigger_frame) #will print the frame number exactly played at 43 seconds
```

```python
# Go to frame selected by timestamp
vid.set(1,trigger_frame) #This will return only 1st frame of the time stamp
success,frame=vid.read()
#Save the frame into an image file in the current working directory
if success:
    cv2.imwrite('screen1.jpg',frame)
```

```python
vid.set(1,trigger_frame+15) #This will return only 15th frame of the time stamp, it can go upto 30
success,frame=vid.read()
#Save the frame into an image file in the current working directory
if success:
    cv2.imwrite('screen15.jpg',frame)
```
